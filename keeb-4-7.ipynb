```python
import cadquery as cq
slant = 12
tent = 27
w = 19 * 5
h = 19 * 3
th = 1.5 # mx plate thickness
front = cq.Workplane('front')
plate = front.box(w, h, th)
bottom = front.box(w*2, h, th).translate((w/2, 0, 0)).translate((0, 0, -70))
plate
```

```python
plate = plate.union(front.box(19*2,19, th).translate((-9.5, -19*2, 0)))
plate
```

```python
keyholes = front.rarray(19, 19, 5, 4).box(14, 14, th*2).translate((0, -9.5, 0))
keyholes
```

```python
plate = plate - keyholes
plate
```

```python
left = plate.rotate((0, 0, 0), (0, 0, 1), -slant)
right = left.mirror(plate.faces(">X")).translate((19/2, 0, 0))
right.union(left)
```

```python
left = left.rotate((w, 0, 0), (w, h, 0), -tent)
right = right.rotate((w, 0, 0), (w, h, 0), tent).translate((0, 0, -39))


thumb1 = front.box(19, 19, th).cut(front.box(14, 14, th))
thumb2 = thumb1.rotate((0, 0, 0), (0, 0, 19), -18).translate((23, -4, 0))
thumb3 = thumb2.rotate((0, 0, 0), (0, 0, 19), -18).translate((23, -4, 0))
thumbs = thumb1.union(thumb2).union(thumb3).rotate((0, 0, 0), (0, 0, 19), -18).translate((19, -19*2, -39))


left.union(right).union(bottom).union(thumbs)
```
